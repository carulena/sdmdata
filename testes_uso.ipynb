```python
import requests

url = "https://specieslink.net/ws/1.0/search?scientificname=Tetragonisca angustula&apikey=2o9Q2mWv9UE8zOSfEffU&coordinates=yes"

payload = {}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```

```python
data = response.json()
all_results = []
for i in data['features']:
    property = i['properties']
    if('daycollected' not in property  or 'monthcollected' not in property ):
        continue
    all_results.append(property)
```

```python
import pandas as pd 

df = pd.DataFrame(all_results)

df.columns
```
